```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob2
from ipywidgets import interact
import ipywidgets as widgets
```

## files env

```python
testDataFolder = "../testdata"
allFiles = glob2.glob(testDataFolder +"/G*"+ "/*.txt") 
```

## load and concate data

```python
tables = dict()
for file in allFiles:
    tables[file] = pd.read_csv(file, sep = "\t", header = [0], skiprows = [2])
    tables[file].drop(tables[file].columns[-1], axis = 1, inplace = True)
```

## plots

```python
fig = plt.figure()
for sensor in tables[allFiles[0]].columns:
    fig = plt.plot(tables[allFiles[0]][sensor], label = sensor)
plt.legend()
```

```python
def TablePlot(tableName):
    fig = plt.figure()
    for sensor in tables[tableName].columns:
        fig = plt.plot(tables[tableName][sensor], label = sensor)
    plt.legend()
```

```python
interact(TablePlot, tableName = allFiles);
```
